```python
import pandas as pd

df = pd.read_csv('data.csv')
```

```python
df
```

```python
symptoms = {
    'Symptom1': {'Fever': ['Mild', 'Low', 'High']},
    'Symptom2': {'Cough': ['Mild', 'Low', 'High']},
    'Symptom3': {'Cold': ['Mild', 'Low', 'High']}
}

print(symptoms)
```

```python
symptom_dict = {}


symptom_dict["Symptoms - 1"] = {
    "Fever": {
        "Mild": df["Fever_Mild"].tolist(),
        "Low": df["Fever_Low"].tolist(),
        "High": df["Fever_High"].tolist()
    }
}


symptom_dict["Symptoms - 2"] = {
    "Cough": {
        "Mild": df["Cough_Mild"].tolist(),
        "Low": df["Cough_Low"].tolist(),
        "High": df["Cough_High"].tolist()
    }
}


symptom_dict["Symptoms - 3"] = {
    "Cold": {
        "Mild": df["Cold_Mild"].tolist(),
        "Low": df["Cold_Low"].tolist(),
        "High": df["Cold_High"].tolist()
    }
}

for i, symptom in enumerate(df["Other_Symptoms"].unique(), 4):
    symptom_dict[f"Symptoms - {i}"] = {
        symptom: ["Low", "Medium", "High"]
    }

print(symptom_dict)
```

```python
import json
```

```python
json_output = json.dumps(symptom_dict, indent=4)


print(json_output)


with open("symptoms_output.json", "w") as json_file:
    json_file.write(json_output)
```
